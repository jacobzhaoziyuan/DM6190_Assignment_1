```python
import sys
sys.path.append('/home/ziyuan/UDA/code_crossmoda/UNet/networks')
from unet import UNet, OUNet
from fcn8s import VGG16_FCN8s as FCN
from drn import drn26 as DRN
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
import cv2
```

```python
def get_parameter_number(net):
    total_num = sum(p.numel() for p in net.parameters())
    trainable_num = sum(p.numel() for p in net.parameters() if p.requires_grad)
    return {'Total': total_num, 'Trainable': trainable_num}
```

```python
model = UNet(in_chns=1, class_num=3)
print('UNet:',get_parameter_number(model))
model = OUNet(in_chns=1, class_num=3)
print('OUNet:',get_parameter_number(model))
model = FCN(num_cls=3, pretrained=True, weights_init=None)
print('FCN:',get_parameter_number(model))
model = DRN(num_cls=3, weights_init=None)
print('DRN:',get_parameter_number(model))
```

```python
vol_ind = 96
slice_ind = 55

f = plt.figure(figsize=(15,4))
model_paths = ['../result/' + model for model in ['DRN','FCN','OUNet','UNet']]
labels=["VS", "Cochlea"]
cp = ["red", "yellow"]
patches = [mpatches.Patch(color=cp[i], label=labels[i]) for i in range(len(cp))]
cmap = matplotlib.colors.ListedColormap(["gray", "red", "yellow"])
plt.legend(handles=patches, ncol=4, fontsize=12, loc="upper center", bbox_to_anchor=(0.5,1))
plt.xticks([]) 
plt.yticks([]) 

ax = 1

img_alpha = 0.99
label_alpha = 0.6
img = np.load(f'/diskd/ziyuan/crossmoda/process/npy/source_training/crossmoda_{vol_ind}_{slice_ind}.npy')
label = np.load(f'/diskd/ziyuan/crossmoda/process/npy/source_training/crossmoda_{vol_ind}_{slice_ind}_label.npy')
f.add_subplot(1,len(model_paths)+1,ax); ax+=1
plt.xticks([]) 
plt.yticks([]) 
plt.xlabel('Ground Truth',fontsize=14)
plt.imshow(img,alpha=img_alpha,cmap='gray')
plt.imshow(label,alpha=label_alpha,vmin=0, vmax=len(cmap.colors),cmap=cmap)

for model_path in model_paths:

    predict = np.load(model_path+f'/{vol_ind}.npy')
    pred = predict[slice_ind]
    f.add_subplot(1,len(model_paths)+1,ax); ax+=1
    plt.xticks([]) 
    plt.yticks([]) 
    model = model_path.split('/')[-1]
    # print(model)
    if model == 'UNet':
        plt.xlabel('Small UNet',fontsize=14)
    elif model == 'OUNet':
        plt.xlabel('UNet',fontsize=14)
    else:
        plt.xlabel(model,fontsize=14)
    
    plt.imshow(img,alpha=img_alpha,cmap='gray')
    plt.imshow(pred.astype(np.uint8), alpha=label_alpha, vmin=0, vmax=len(cmap.colors),cmap=cmap)

plt.savefig('source training_1.png')
plt.show()
```
